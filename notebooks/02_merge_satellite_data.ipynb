```python
import pandas as pd

optical = pd.read_csv('/content/cleaned_satellite_dataset.csv')

sar = pd.read_csv('/content/AgriAstra_Sentinel1.csv')
```

```python
sar.head()
```

```python
sar.columns
```

```python
optical = optical.drop(columns=['.geo'])

sar = sar.drop(columns=['system:index', '.geo'])

optical.head()

sar.head()
```

```python
print(optical.shape)

print(sar.shape)
```

```python
master_df = pd.concat([optical, sar], axis=1)

master_df.head()
```

```python
master_df.columns
```

```python
master_df.to_csv(
'merged_satellite_dataset.csv',
index=False
)

print("Master dataset created.")
```
