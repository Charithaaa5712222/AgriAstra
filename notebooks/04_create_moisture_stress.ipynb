```python
import pandas as pd

df = pd.read_csv('/content/final_agriastra_dataset.csv')

df.head()
```

```python
def classify_stress(row):

    if row['NDWI'] < -0.6:
        return 'High'

    elif row['NDWI'] < -0.3:
        return 'Moderate'

    else:
        return 'Low'


df['Moisture_Stress'] = df.apply(
    classify_stress,
    axis=1
)

df.head()
```

```python
df['Moisture_Stress'].value_counts()
```

```python
df.to_csv(
    'agriastra_ai_dataset.csv',
    index=False
)

print("AI dataset ready.")
```
